## **ğŸ Data Analysis 3ë‹¨ê³„**

```
'Class' ì´ìƒì—¬ë¶€ë¥¼ íŒë‹¨í•  ìˆ˜ ìˆëŠ” MLì•Œê³ ë¦¬ì¦˜ í•™ìŠµ/ì˜ˆì¸¡ í‰ê°€
"Q. ì–´ë–»ê²Œ Classê°€ 1ì¸ ë°ì´í„°ë¥¼ ì¶”ì¶œí•  ìˆ˜ ìˆì„ê¹Œ?"

â€» ìˆ˜ê°•ìƒ í•™ìŠµ ë° ì ìš©í•´ì•¼í•˜ëŠ” ì‚¬í•­ (*ì¶”ê°€ í•™ìŠµ ìë£Œ í™œìš©)

 - Feature Selection / Feature Extraction ê°œë… ì´í•´ ë° ì ìš©í•´ë³´ê¸°
```  

````python
'''
1. PCA + Mahalanobis Distance

í•´ë‹¹ ì›¨ì´í¼ ë°ì´í„°ì˜ ê²½ìš° íƒ€ê²Ÿ ë³€ìˆ˜ë¥¼ ì œì™¸í•œ 1558 ì°¨ì›ì˜ ë§¤ìš° ê³ ì°¨ì› ë°ì´í„°ì´ë¯€ë¡œ, ì°¨ì› ì¶•ì†Œë¥¼ í•  í•„ìš”ê°€ ìˆìŒ.
ëŒ€í‘œì ì¸ ì°¨ì› ì¶•ì†Œ ê¸°ë²•ì¸ PCA ê¸°ë²• í™œìš© + Mahalanobis Distance ì´ìš©í•œ ê²°ê³¼ ë¹„êµ
'''

# 1) PCA í›„ ë³€ìˆ˜ ì„¤ëª…ë ¥

from sklearn.decomposition import PCA
import time

pca = PCA()

reduced_ss = pca.fit_transform(X_ss)

# â–¶ pca.n_components_(ì°¨ì› ì¶•ì†Œ ì£¼ì„±ë¶„ ê°œìˆ˜)
features = range(pca.n_components_)
feature_df=pd.DataFrame(data=features,columns=['pc_feature'])

# â–¶ pca.explained_variance_ratio_(ì„¤ëª…ë ¥)
variance_df=pd.DataFrame(data=pca.explained_variance_ratio_,columns=['variance'])
pc_feature_df=pd.concat([feature_df,variance_df],axis=1)
pc_feature_df
````

````python
# 400ê°œ ì£¼ì„±ë¶„ìœ¼ë¡œ ê±°ì˜ 100% ì„¤ëª…ê°€ëŠ¥
fig, ax = plt.subplots(figsize = (10,4))
xi = np.arange(1, reduced_ss.shape[1]+1, step=1)
y = np.cumsum(pca.explained_variance_ratio_)*100

plt.ylim(0.0,101)
plt.xlabel('number of principle components')
plt.ylabel('explained features (%)')
sns.scatterplot(x=xi, y=y)
plt.show()
````

````python
# 400ê°œ ì£¼ì„±ë¶„ ì‚¬ìš©ìœ„í•œ ì°¨ì› ì¶•ì†Œ ì‹œí–‰

pca = PCA(n_components=400)
pc_ss = pca.fit_transform(X_ss)

pc_ss = pd.DataFrame(pc_ss).reset_index(drop=True)
pc_ss.head()
````

````python
# Mahalanobis Distance ê³„ì‚° ìœ„í•œ ëª¨ë¸ ì„ ì •. ì—¬ê¸°ì„œëŠ” pyodë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ MCD ëª¨ë¸ ì‚¬ìš©
!pip install pyod
````

````python
# ê°€ì§„ ë°ì´í„°ì˜ ì´ìƒì¹˜ ë¹„ìœ¨ì„ contaminationìœ¼ë¡œ ì„¤ì •

from pyod.models.mcd import MCD

t0 = time.time()
outlier_detector = MCD(contamination = df_ss['Class'].value_counts(normalize=True).values[1], random_state=42)
outlier_detector.fit(pc_ss)

# outlier ê²€ì¶œ
y_pred = outlier_detector.predict(pc_ss)
t1 = time.time()

print(y_pred)
print('\n')
print('------------------------------')
print('\n')
print(f'The MCD model training took {t1 - t0} seconds')
````

````python
pd.Series(y_pred).value_counts()
````

````python
# ì„±ëŠ¥í‰ê°€
from sklearn.metrics import classification_report
print(classification_report(Y, y_pred))
````

````python
# Sklearnì˜ ElipticEnvelopeì™€ ë¹„êµ
from sklearn.covariance import EllipticEnvelope

t2 = time.time()
outlier_detector_2 = EllipticEnvelope(contamination=df_ss['Class'].value_counts(normalize=True).values[1], random_state=42)

outlier_detector_2.fit(pc_ss)

pred_2 = outlier_detector_2.predict(pc_ss)
y_pred_2 = np.where(pred_2==-1, 1, 0)

t3 = time.time()

print(y_pred_2)
print('\n')
print('------------------------------')
print('\n')
print(f'The EllipticEnvelope model training took {t3 - t2} seconds')
````

````python
'''
PCA + Mahalanobis ë°©ì‹ì˜ ê²½ìš°, ì°¨ì› ì¶•ì†Œë¥¼ í–ˆìŒì—ë„ ë¶ˆêµ¬í•˜ê³ 
400ê°œì˜ ì°¨ì›ì„ ì´ìš©í•´ì•¼í•˜ë©° PCAì‹œê°„ì„ ì œì™¸í–ˆìŒì—ë„ ë¶ˆêµ¬í•˜ê³  í•™ìŠµ ì‹œê°„ì´ ì•½ 2ë¶„ìœ¼ë¡œ ê½¤ ì˜¤ë˜ê±¸ë¦¼.
ë˜í•œ Mahalanobis Distanceì˜ ê²½ìš° ê³µë¶„ì‚° ê°œë…ì„ ì´ìš©í•˜ëŠ”ë°, PCAë¡œ ì°¨ì›ì¶•ì†Œí•œ ë³€ìˆ˜ë“¤ì€ ê°ê° ìˆ˜ì§ìœ¼ë¡œ ê³µë¶„ì‚°ì´ 0ì„.
ë”°ë¼ì„œ 1ìˆœìœ„ ë°©ì‹ì„ ë³´ë¥˜í•˜ê³ , 2ìˆœìœ„ ë°©ì‹ì„ ê³ ë ¤

2. IsolationForest : Treeëª¨ë¸ ê¸°ë°˜ ì´ìƒíƒì§€ ì‹œí–‰
'''

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_ss, Y, test_size=0.3, stratify=Y, random_state=42)

print(f'x_train shape : {x_train.shape}')
print(f'x_test shape : {x_test.shape}')
print(f'y_train shape : {y_train.shape}')
print(f'y_test shape : {y_test.shape}')
````

````python
pd.Series(y_train).value_counts(normalize=True)
````

````python
pd.Series(y_test).value_counts(normalize=True)
````

````python
from sklearn.ensemble import IsolationForest

t5 = time.time()
clf = IsolationForest(max_samples=200, contamination = df_ss['Class'].value_counts(normalize=True).values[1], random_state=42)
clf.fit(x_train)

y_pred_train = clf.predict(x_train)
y_pred_test = clf.predict(x_test)

t6 = time.time()
y_pred_train = np.where(y_pred_train==-1, 1, 0)
y_pred_test = np.where(y_pred_test==-1, 1, 0)

print(classification_report(y_train, y_pred_train))
print('\n')
print('-----------------------------------------------------')
print('\n')
print(classification_report(y_test, y_pred_test))
print('\n')
print('-----------------------------------------------------')
print(f'The IsolationForest model training took {t6 - t5} Seconds')
````

````python
# Isolation Forest ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ì¡°ê¸ˆ ë” ì˜¬ë ¤ë³´ê¸° ìœ„í•´, scoringê¸°ë°˜ìœ¼ë¡œ Threshold ì¡°ì •
y_pred_train[10:20], clf.decision_function(x_train)[10:20]
````

````python
# Score ë³€ìˆ˜ í• ë‹¹
y_pred_train_score =  clf.decision_function(x_train)
y_pred_test_score = clf.decision_function(x_test)

# Threshold ì¡°ì •
y_pred_train = np.where(y_pred_train_score < 0.006, 1, 0)
y_pred_test = np.where(y_pred_test_score < 0.006, 1, 0)
````
