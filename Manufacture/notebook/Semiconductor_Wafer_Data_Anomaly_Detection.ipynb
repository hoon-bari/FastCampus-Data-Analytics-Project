````python
# 이상 비율 확인하기 -> 8.34%
# 그 전에, Class를 정수로 바꿔줌
df['Class'] = df['Class'].astype(int)
print(round(df['Class'].value_counts(normalize = True) * 100, 2))
````

````python
# 이상치 데이터와 normal 데이터 구분
df_normal = df[df['Class'] == 0]
df_abnormal = df[df['Class'] == 1]
````

````python
# 각 feature의 평균 확인
df_normal_mean = pd.DataFrame(df_normal.describe()).loc['mean']
df_abnormal_mean = pd.DataFrame(df_abnormal.describe()).loc['mean']
````

````python
# 1. gap을 통해, 어떤 피쳐가 차이가 많이나면 이상치를 가늠할 수 있는 가늠자로 판단할 수 있음
# 2. 하지만 단위가 다를 경우, 단위에 의해 갭 차이가 더 부각될 수 있음. 표준화를 통해 각 피쳐 값 조정 후 재확인 필요.
df_concat = pd.concat([df_normal_mean, df_abnormal_mean], axis=1, keys=['normal', 'abnormal'])
df_concat['gap'] = abs(df_concat['normal'] - df_concat['abnormal'])
df_concat.head()
````

````python
# 갭 기준 내림차순 정렬 -> 단위 차이에 의해 feature 2의 갭이 부각됨
df_gap_10 = df_concat.sort_values(by=['gap'], ascending=False).drop('Class').head(10)
df_gap_10[['gap']]
````

````python
# 표준화 후 gap 재확인
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X = df.drop(['Class'], axis=1)
Y = df['Class']

X_ss = ss.fit_transform(X)

df_ss = pd.concat([pd.DataFrame(X_ss, columns=X.columns), Y], axis=1)
df_ss
````

````python
# 표준화 후 gap을 다시 비교해보면, feature_2가 아닌 feature_1400이 gap 1순위가 됨
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use(['dark_background'])

df_ss_normal = df_ss[df_ss['Class']==0]
df_ss_abnormal = df_ss[df_ss['Class']==1]

df_ss_normal_mean = pd.DataFrame(df_ss_normal.describe()).loc['mean']
df_ss_abnormal_mean = pd.DataFrame(df_ss_abnormal.describe()).loc['mean']

df_ss_concat = pd.concat([df_ss_normal_mean, df_ss_abnormal_mean], axis=1, keys=['normal', 'abnormal'])
df_ss_concat['gap'] = abs(df_ss_concat['normal'] - df_ss_concat['abnormal'])

df_ss_top10 = df_ss_concat.sort_values(by=['gap'], ascending=False).drop('Class').head(10)

print(df_ss_top10)
print('\n')

fig = plt.figure(figsize=(15,5));
plt.title('Features Gap');
plt.bar(df_ss_top10.index, df_ss_top10['gap']);
````

````python
'''
1. PCA + Mahalanobis Distance

해당 웨이퍼 데이터의 경우 타겟 변수를 제외한 1558 차원의 매우 고차원 데이터이므로, 차원 축소를 할 필요가 있음.
대표적인 차원 축소 기법인 PCA 기법 활용 + Mahalanobis Distance 이용한 결과 비교
'''

# 1) PCA 후 변수 설명력

from sklearn.decomposition import PCA
import time

pca = PCA()

reduced_ss = pca.fit_transform(X_ss)

# ▶ pca.n_components_(차원 축소 주성분 개수)
features = range(pca.n_components_)
feature_df=pd.DataFrame(data=features,columns=['pc_feature'])

# ▶ pca.explained_variance_ratio_(설명력)
variance_df=pd.DataFrame(data=pca.explained_variance_ratio_,columns=['variance'])
pc_feature_df=pd.concat([feature_df,variance_df],axis=1)
pc_feature_df
````

````python
# 400개 주성분으로 거의 100% 설명가능
fig, ax = plt.subplots(figsize = (10,4))
xi = np.arange(1, reduced_ss.shape[1]+1, step=1)
y = np.cumsum(pca.explained_variance_ratio_)*100

plt.ylim(0.0,101)
plt.xlabel('number of principle components')
plt.ylabel('explained features (%)')
sns.scatterplot(x=xi, y=y)
plt.show()
````

````python
# 400개 주성분 사용위한 차원 축소 시행

pca = PCA(n_components=400)
pc_ss = pca.fit_transform(X_ss)

pc_ss = pd.DataFrame(pc_ss).reset_index(drop=True)
pc_ss.head()
````

````python
# Mahalanobis Distance 계산 위한 모델 선정. 여기서는 pyod라이브러리의 MCD 모델 사용
!pip install pyod
````

````python
from sklearn.ensemble import IsolationForest

t5 = time.time()
clf = IsolationForest(max_samples=200, contamination = df_ss['Class'].value_counts(normalize=True).values[1], random_state=42)
clf.fit(x_train)

y_pred_train = clf.predict(x_train)
y_pred_test = clf.predict(x_test)

t6 = time.time()
y_pred_train = np.where(y_pred_train==-1, 1, 0)
y_pred_test = np.where(y_pred_test==-1, 1, 0)

print(classification_report(y_train, y_pred_train))
print('\n')
print('-----------------------------------------------------')
print('\n')
print(classification_report(y_test, y_pred_test))
print('\n')
print('-----------------------------------------------------')
print(f'The IsolationForest model training took {t6 - t5} Seconds')
````

````python
'''
feature importance 확인을 위해 SHAP Value로 시각화

Feature importance의 경우 피처의 변화가 예측에 주는 영향력을 계산함으로써 피처 영향력을 측정하지만,
퍼뮤테이션의 정도와 에러에 기반한 추정 한계 때문에 알고리즘 실행 시마다 중요도가 다를 수 있다.
또한, 피처 중요도는 ‘피처 간 의존성’을 간과한다(피처 간 상관관계가 존재하는 모델은 사용을 지양해야 한다.
(출처 : https://velog.io/@sjinu/%EA%B0%9C%EB%85%90%EC%A0%95%EB%A6%ACSHAPShapley-Additive-exPlanations)

SHAP는 이와 다르게 피처간 의존성까지 고려하여 모델 영향력을 계산한다.
'''

!pip install shap
````

````python
'''
Shap에 따라 나온 feature의 모델에 대한 영향력.
양수 SHAP 값은 해당 feature의 값이 증가할수록 모델의 예측값이 증가한다는 것을 의미.
이는 모델이 해당 feature를 중요하게 여기고 있으며,
이 feature의 값이 증가할 때 모델의 예측값도 함께 증가한다는 것을 나타냄.

음수 SHAP 값은 해당 feature의 값이 증가할수록 모델의 예측값이 감소한다는 것을 의미.
이는 모델이 해당 feature를 부정적인 영향을 미치는 중요한 요소로 인식하고 있으며,
이 feature의 값이 증가할 때 모델의 예측값은 감소한다는 것을 나타냄.

어떤 인자가 SHAP Value가 0이 되는 지점을 좌우로 하여 색깔이 나뉘면 해당 변수는 특정 방향으로 예측값에 영향을 미쳤다는 것이므로,
그 인자는 우선적으로 살필 필요.
'''

import shap

X_explain = x_test
shap_values = shap.TreeExplainer(clf).shap_values(X_explain)
shap.summary_plot(shap_values, X_explain)
````
